## plothrcmodes
Plot reconstructed harmonic components alongside reference modes.

Displays an 8-panel figure (4 rows × 2 columns) comparing the DAHD
reconstructed modes R1–R4 (right column) against the ground-truth
reference modes xref1–xref4 (left column). Reproduces Fig. 5 from
Kondrashov (2026).

**Requires** `generate_data` and `dahd4example` to have been run first,
so that `xref1`–`xref4` and `R1`–`R4` are available in the namespace.

**Dependencies:** `generate_data`, `dahd4example`

```python
xmax = 0.5
xmin = -0.5
```

```python
fig = plt.figure(figsize=(20 / 2.54, 40 / 2.54), dpi=100)
plt.rcParams.update({
    'axes.labelweight': 'bold',
    'axes.titleweight': 'bold',
    'axes.labelsize': 8,
    'axes.titlesize': 10,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
})
```

-- Reconstructed components --

```python
plt.subplot(422)
cf = plt.contourf(R1.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Recons at f_1')
plt.xlim([1, 129])
```

```python
plt.subplot(424)
cf = plt.contourf(R2.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Recons at f_2')
plt.xlim([1, 129])
```

```python
plt.subplot(426)
cf = plt.contourf(R3.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Recons at f_3')
plt.xlim([1, 129])
```

```python
plt.subplot(428)
cf = plt.contourf(R4.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Recons at f_4')
plt.xlabel('Time')
plt.xlim([1, 129])
```

-- Reference modes --

```python
plt.subplot(421)
cf = plt.contourf(xref1.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Mode at f_1')
plt.xlim([1, 129])
```

```python
plt.subplot(423)
cf = plt.contourf(xref2.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Mode at f_2')
plt.xlim([1, 129])
```

```python
plt.subplot(425)
cf = plt.contourf(xref3.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Mode at f_3')
plt.xlim([1, 129])
```

```python
plt.subplot(427)
cf = plt.contourf(xref4.T, 20, cmap='jet', vmin=xmin, vmax=xmax)
plt.colorbar(cf)
plt.title('Mode at f_4')
plt.xlabel('Time')
plt.xlim([1, 129])
```

```python
plt.tight_layout()
plt.show()
```
